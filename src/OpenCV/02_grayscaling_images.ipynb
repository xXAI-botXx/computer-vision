# Grayscaling Images

- Apply grayscaling
- Dimensions differences

---
### Imports

```python
import matplotlib.pyplot as plt
import cv2
from urllib.request import urlretrieve
```

```python
def imshow(img, title=None, size=10, axis=False):
    aspect_ratio = img.shape[0] / img.shape[1]
    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    if type(title) == str:
        plt.title(title)
    if axis ==False:
        plt.axis('off')
    plt.show()

def get_img_dim(img):
    res = "Image Dimensions:\n"
    res += f"    -> Height: {int(img.shape[0])} pixels\n"
    res += f"    -> Width: {int(img.shape[1])} pixels\n"
    if len(img.shape) > 2:
        res += f"    -> Color Channels: {int(img.shape[2])} components\n"
    print(res)
    return img.shape
```

---
### Load Image

```python
url = 'https://upload.wikimedia.org/wikipedia/commons/0/0e/Changchun_Montage_2017.png'
file_path = '../../res/img/changchun.jpg'
urlretrieve(url, file_path)

img = cv2.imread(file_path)
imshow(img, "Changchun", size=8, axis=False)
```

---
### Apply grayscale

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imshow(gray_img, "Gray Image")
```

---
### Dimension differences

Grayscale images are just 2 Dimensions. Height and Width position and the intensity value.

```python
get_img_dim(gray_img)
```

```python
get_img_dim(img)
```
