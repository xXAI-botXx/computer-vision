### Displaying images

Matplotlib and OpenCV uses a different order of RGB, so we have to change the order of the channels, when displaying the img.

Matplotlib: RGB<br>
OpenCV: BGR

````python
import matplotlib.pyplot as plt

print("Image with wrong RGB Channels")
plt.imshow(img);
````

````python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));
````

````python
red_img = np.copy(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
red_img[:, :, 1] = 0
red_img[:, :, 2] = 0
plt.imshow(red_img, cmap="gray")
plt.show()

green_img = np.copy(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
green_img[:, :, 0] = 0
green_img[:, :, 2] = 0
plt.imshow(green_img, cmap="gray")
plt.show()

blue_img = np.copy(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
blue_img[:, :, 0] = 0
blue_img[:, :, 1] = 0
plt.imshow(blue_img, cmap="gray");
````

Wrapper Function for plotting

````python
def imshow(img, title=None):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    if type(title) == str:
        plt.title(title)
    plt.show()

imshow(img)
imshow(img, "NVIDIA Tesla GPU")
````

````python
def get_img_dim(img):
    res = "Image Dimensions:\n"
    res += f"    -> Height: {int(img.shape[0])} pixels\n"
    res += f"    -> Width: {int(img.shape[1])} pixels\n"
    res += f"    -> Color Channels: {int(img.shape[2])} components\n"
    print(res)
    return img.shape

get_img_dim(img)
````
